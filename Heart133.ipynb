**Correlation Matrix**

A correlation matrix is a table that displays the correlation coefficients between pairs of variables in a dataset. It helps to understand how different variables are related to each other. Each cell in the matrix represents the correlation between two variables.

```python
# Compute the correlation matrix
corr_matrix = df.corr()

# Display the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```

*df.corr()*:It computes the correlation matrix for numerical columns in the DataFrame.

*seaborn.heatmap()*:It is used to visualize the correlation matrix as a heatmap, where the colors help quickly identify strong positive or negative correlations.

If we observe the output of above code which is correlation matrix You might see how features like age, cholesterol levels, exercise, and blood pressure are correlated with each other and with the target variable which is presence of heart disease or not.

**Inferential Statistics**

*Hypothesis Testing*:
Let's test whether the average maximum heart rate achieved thalach differs significantly between patients with and without heart disease.

*Null Hypothesis (H₀)*: There is no difference in average thalach between patients with and without heart disease.

*Alternative Hypothesis (H₁)*: There is a significant difference in average thalach between the two groups.

```python
# Separate the data into two groups
thalach_disease = df[df['target'] == 1]['thalach']
thalach_no_disease = df[df['target'] == 0]['thalach']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(thalach_disease, thalach_no_disease)

print("T-Statistic:",t_stat)
print("P-Value:",p_value)
```

If the p-value is less than 0.05, we reject the null hypothesis therefore, There is a significant difference in average thalach between the two groups.
